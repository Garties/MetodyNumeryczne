**Zadanie 1.** 

Rozważmy funkcję $f(x)=\sqrt{\operatorname{tg}^2(x)+1}-\operatorname{tg}(x)$. Zaimplementuj ją w Pythonie oraz narysuj jej wykres w przedziale $[-2\pi, 2\pi]$.

```python
def f(x):
    return np.sqrt(np.tan(x)**2 + 1) - np.tan(x)

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$f(x)=\sqrt{\operatorname{tg}^2(x)+1}-\operatorname{tg}(x)$')

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'Wykres funkcji $f(x)=\sqrt{\operatorname{tg}^2(x)+1}-\operatorname{tg}(x)$ w przedziale $[-2\pi, 2\pi]$')
plt.legend()

plt.ylim(-0.5, 1.5)

plt.grid(True)

plt.show()
```

**Zadanie 2.**

Zaimplementuj wybrane metody różniczkowania numerycznego (co najmniej 4) przedstawione na wykładzie i porównaj dokładność wyznaczania za ich pomocą pochodnej funkcji z zadania 1. Skoncentruj się na pochodnych dla wartości leżących w pobliżu punktu $\frac{\pi}{2}+k\pi$. W porównaniu wykorzystaj błąd względy. Za wartość dokładną możesz przyjąć wartość pochodnej $f(x)$ obliczoną analitycznie ($f^{\prime}(x)=\operatorname{tg}(x) \sqrt{\operatorname{tg}^2(x)+1}-\left(\operatorname{tg}^2(x)+1\right)$) lub wyznaczoną za pomocą pakietu SymPy
.

```python
def f(x):
    return np.sqrt(np.tan(x)**2 + 1) - np.tan(x)

def exact_derivative(x):
    return np.tan(x) * np.sqrt(np.tan(x)**2 + 1) - (np.tan(x)**2 + 1)

def forward_difference(f, x, h):
    return (f(x + h) - f(x)) / h

def backward_difference(f, x, h):
    return (f(x) - f(x - h)) / h

def central_difference(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

def second_central_difference(f, x, h):
    return (-f(x + 2*h) + 8*f(x + h) - 8*f(x - h) + f(x - 2*h)) / (12 * h)

points = [np.pi/2 + k*np.pi for k in range(-2, 3)]

h = 1e-5

results = {}
for point in points:
    if point % np.pi != np.pi/2:
        results[point] = {
            'exact': exact_derivative(point),
            'forward': forward_difference(f, point, h),
            'backward': backward_difference(f, point, h),
            'central': central_difference(f, point, h),
            'second_central': second_central_difference(f, point, h)
        }

errors = {}
for point, values in results.items():
    exact = values['exact']
    errors[point] = {
        'forward': np.abs(values['forward'] - exact) / np.abs(exact),
        'backward': np.abs(values['backward'] - exact) / np.abs(exact),
        'central': np.abs(values['central'] - exact) / np.abs(exact),
        'second_central': np.abs(values['second_central'] - exact) / np.abs(exact)
    }

errors
```
